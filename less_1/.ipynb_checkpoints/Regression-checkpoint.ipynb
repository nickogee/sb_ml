```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
matplotlib.rc("font", size=18)
```

```python
houses = pd.read_csv("1.4_houses.csv")
houses.head(10)
```

Посмотрим как соотносятся  dim_1 dim_2 с цкнами

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()

ax.scatter(houses["dim_1"], houses["price"], s=100)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# выгружаем признаки и целевые значения в отдельные переменные
X = houses[['dim_1']]
y = houses['price']

X
```

```python
#создаем регрессор
reg = LinearRegression().fit(X, y)
```

```python
# price = f(x) = a + b*x
```

```python
# тогда константа a:
a = reg.intercept_
a
```

```python
# и коэффециент b
[b] = reg.coef_
[b]
```

Тогда можем написать функцию для вычисления цены дома

```python
def reg_prediction(dim_1):
    return a + b * dim_1
```

```python
# предсказанная цена для первого дома:
reg_prediction(X.dim_1[0])
```

```python
# а его реальная цена из данных
y[0]
```

Используем встроенные методы для тех-же вычислений

```python
X[0:1]
```

```python
reg.predict(X[0:1])
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()

# координаты данных
ax.scatter(houses["dim_1"], houses["price"], s=100)

# предсказания
ax.plot([X.dim_1.min(), X.dim_1.max()], [reg_prediction(X.dim_1.min()), reg_prediction(X.dim_1.max())], c='red')

plt.show()
```

### Поэксперементируем с двумя измерениями для оценки стоимости дома

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

# координаты данных
ax.scatter(houses["dim_1"], houses["dim_2"], houses["price"], s=100)

ax.set_xlabel("dim_1")
ax.set_ylabel("dim_2")
ax.set_zlabel("price")

plt.show()
```

```python
# выгружаем признаки и целевые значения в отдельные переменные
X = houses[['dim_1', 'dim_2']]
y = houses['price']
```

```python
#создаем регрессор
reg = LinearRegression().fit(X, y)
```

```python
# теперь функция для нахождения целевой переменной (цены):
# price = f(x1, x2) = a + b1*x1 + b2*x2

[b1, b2] = reg.coef_
a = reg.intercept_

def reg_prediction(dim_1, dim_2):
    return a + b1 * dim_1 + b2 * dim_2
```

```python
# предсказанная цена для первого дома:
reg_prediction(X.dim_1[0], X.dim_2[0])
```

```python
# используем встроенные методы
reg.predict(X[0:1])
```

```python
# чтобы нарисовать плоскость (в 3d) с предсказаниями цены домов,
# сделаем предсказания для различных конфигураций домов

d1, d2 = list(), list()

for x in np.linspace(min(houses["dim_1"]), max(houses["dim_1"]), 100):
    for y in np.linspace(min(houses["dim_2"]), max(houses["dim_2"]), 100):
        d1.append(x)
        d2.append(y)

d1 = np.array(d1).reshape(-1, 1)
d2 = np.array(d2).reshape(-1, 1)
```

```python

np.concatenate([d1, d2], axis=1)
```

```python
p = reg.predict(np.concatenate([d1, d2], axis=1))
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

# координаты данных
ax.scatter(houses["dim_1"], houses["dim_2"], houses["price"], s=100)

# помещаем плоскость предсказанных значений
ax.plot_trisurf(d1.ravel(), d2.ravel(), p.ravel(), alpha=0.2)

ax.set_xlabel("dim_1")
ax.set_ylabel("dim_2")
ax.set_zlabel("price")

# изменим позицию камеры
ax.elev = 12

plt.show()
```
