```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

matplotlib.rc("font", size=18)
```

```python
houses = pd.read_csv("1.4_houses.csv")
houses.head(10)
```

```python
fig = plt.figure(figsize=(5, 5))

markers = {
    'basic':'o',
    'medium': 'v',
    'luxury': '*'
}

for d1, d2, l in zip(houses.dim_1, houses.dim_2, houses.level):
    plt.scatter(d1, d2, s=100, marker=markers[l])

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# выгружаем признаки в отдельную переменную
X = houses[['dim_1', 'dim_2']]
```

```python
# Создаем модель для кластеризации
clust = KMeans(n_clusters=3).fit(X)
```

```python
# центры получившихся кластеров
c1, c2, c3 = clust.cluster_centers_
```

```python
# Посмотрим расположение центров кластеров

fig = plt.figure(figsize=(7, 7))

markers = {
    'basic':'o',
    'medium': 'v',
    'luxury': '*'
}

for d1, d2, l in zip(houses.dim_1, houses.dim_2, houses.level):
    plt.scatter(d1, d2, s=100, marker=markers[l])

plt.scatter(c1[0], c1[1], s=100, marker='x', c='black') 
plt.scatter(c1[0], c1[1], s=250*0.8e2, c='black', alpha=0.1)

plt.scatter(c2[0], c2[1], s=100, marker='x', c='black') 
plt.scatter(c2[0], c2[1], s=250*0.7e2, c='black', alpha=0.1)

plt.scatter(c3[0], c3[1], s=100, marker='x', c='black') 
plt.scatter(c3[0], c3[1], s=250*3e2, c='black', alpha=0.1)

    
plt.show()
```
